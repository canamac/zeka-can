```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Coin listesi (BTCUSDT ve diğer 10 coin)
coins = [
    "BTC-USD",  # BTCUSDT
    "ETH-USD",  # Ethereum
    "BNB-USD",  # Binance Coin
    "ADA-USD",  # Cardano
    "DOGE-USD", # Dogecoin
    "XRP-USD",  # Ripple
    "DOT-USD",  # Polkadot
    "UNI-USD",  # Uniswap
    "LTC-USD",  # Litecoin
    "LINK-USD", # Chainlink
    "BCH-USD"   # Bitcoin Cash
]

# Geçmiş verileri çekme (1 yıl)
data = yf.download(coins, start="2022-01-01", end="2024-01-31")['Close']

# BTCUSDT'ye göre göreceli performansı hesapla
btc_prices = data["BTC-USD"]
relative_performance = data.div(btc_prices, axis=0)

# Performansı normalize et (başlangıç değeri = 1)
normalized_performance = relative_performance / relative_performance.iloc[0]

# Grafik çizimi
plt.figure(figsize=(12, 6))
for coin in coins[1:]:  # BTC hariç diğer coinler
    plt.plot(normalized_performance.index, normalized_performance[coin], label=coin)

plt.axhline(1, color='black', linestyle='--', label="BTC-USD")  # BTC referans çizgisi
plt.title("Coinlerin BTCUSDT'ye Göre Göreceli Performansı (2022)")
plt.xlabel("Tarih")
plt.ylabel("Göreceli Performans (Normalize Edilmiş)")
plt.legend()
plt.grid()
plt.show()
```
