```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Coin listesi (BTCUSDT ve dier 10 coin)
coins = [
    "BTC-USD",  # BTCUSDT
    "ETH-USD",  # Ethereum
    "BNB-USD",  # Binance Coin
    "ADA-USD",  # Cardano
    "DOGE-USD", # Dogecoin
    "XRP-USD",  # Ripple
    "DOT-USD",  # Polkadot
    "UNI-USD",  # Uniswap
    "LTC-USD",  # Litecoin
    "LINK-USD", # Chainlink
    "BCH-USD"   # Bitcoin Cash
]

# Gemi verileri ekme (1 yl)
data = yf.download(coins, start="2022-01-01", end="2024-01-31")['Close']

# BTCUSDT'ye gre greceli performans hesapla
btc_prices = data["BTC-USD"]
relative_performance = data.div(btc_prices, axis=0)

# Performans normalize et (balang deeri = 1)
normalized_performance = relative_performance / relative_performance.iloc[0]

# Grafik izimi
plt.figure(figsize=(12, 6))
for coin in coins[1:]:  # BTC hari dier coinler
    plt.plot(normalized_performance.index, normalized_performance[coin], label=coin)

plt.axhline(1, color='black', linestyle='--', label="BTC-USD")  # BTC referans izgisi
plt.title("Coinlerin BTCUSDT'ye Gre Greceli Performans (2022)")
plt.xlabel("Tarih")
plt.ylabel("Greceli Performans (Normalize Edilmi)")
plt.legend()
plt.grid()
plt.show()
```
