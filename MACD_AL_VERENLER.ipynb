```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# BIST 100 hisselerinin sembollerini bir liste olarak tanmlyoruz
bist100_hisseleri = [
    'AEFES.IS', 'AKBNK.IS', 'AKSA.IS', 'ALARK.IS', 'ARCLK.IS', 'ASELS.IS', 'BERA.IS', 'BIMAS.IS', 'DOHOL.IS',
    'EGEEN.IS', 'EKGYO.IS', 'ENJSA.IS', 'ENKAI.IS', 'EREGL.IS', 'FROTO.IS', 'GARAN.IS', 'GUBRF.IS', 'HALKB.IS',
    'HEKTS.IS', 'ISCTR.IS', 'KCHOL.IS', 'KONTR.IS', 'KORDS.IS', 'KOZAA.IS', 'KOZAL.IS', 'KRDMD.IS', 'MGROS.IS',
    'ODAS.IS', 'OYAKC.IS', 'PETKM.IS', 'PGSUS.IS', 'SAHOL.IS', 'SASA.IS', 'SISE.IS', 'SKBNK.IS', 'SOKM.IS',
    'TAVHL.IS', 'TCELL.IS', 'THYAO.IS', 'TKFEN.IS', 'TOASO.IS', 'TSKB.IS', 'TTKOM.IS', 'TUPRS.IS', 'VAKBN.IS',
    'VESTL.IS', 'YKBNK.IS'
]

# MACD al sinyali veren hisseleri saklamak iin bir liste oluturuyoruz
al_sinyali_veren_hisseler = []

# Her bir hisse iin MACD hesaplamalar yapyoruz
for sembol in bist100_hisseleri:
    print(f"{sembol} verileri inceleniyor...")
    try:
        # Hisse verilerini indiriyoruz
        veri = yf.download(sembol, period='1y')

        # MACD hesaplamalar
        veri['EMA12'] = veri['Close'].ewm(span=12, adjust=False).mean()
        veri['EMA26'] = veri['Close'].ewm(span=26, adjust=False).mean()
        veri['MACD'] = veri['EMA12'] - veri['EMA26']
        veri['Sinyal'] = veri['MACD'].ewm(span=9, adjust=False).mean()

        # Son iki gnn MACD ve Sinyal deerlerini alyoruz
        macd_son = veri['MACD'].iloc[-1]
        sinyal_son = veri['Sinyal'].iloc[-1]
        macd_onceki = veri['MACD'].iloc[-2]
        sinyal_onceki = veri['Sinyal'].iloc[-2]

        # MACD'nin Sinyal hattn yukar doru kestiini kontrol ediyoruz
        if macd_onceki < sinyal_onceki and macd_son > sinyal_son:
            al_sinyali_veren_hisseler.append(sembol)
            print(f"{sembol}: MACD AL sinyali verdi!")

            # Grafik izimi
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [3, 1]})

            # st grafik: Kapan fiyatlar
            ax1.plot(veri.index, veri['Close'], label='Kapan Fiyat', color='blue')
            ax1.set_title(f'{sembol} Kapan Fiyat ve MACD AL Sinyali')
            ax1.set_ylabel('Fiyat (TRY)')
            ax1.legend()
            ax1.grid(True)

            # Alt grafik: MACD gstergesi
            ax2.plot(veri.index, veri['MACD'], label='MACD', color='green')
            ax2.plot(veri.index, veri['Sinyal'], label='Sinyal Hatt', color='red')
            ax2.bar(veri.index, veri['MACD'] - veri['Sinyal'], label='MACD Histogram', color='grey', alpha=0.3)

            ax2.set_title('MACD Gstergesi')
            ax2.set_ylabel('Deer')
            ax2.legend()
            ax2.grid(True)

            # Grafikleri dzenle ve gster
            plt.tight_layout()
            plt.show()

    except Exception as e:
        print(f"{sembol} verileri alnrken hata olutu: {e}")

# MACD al sinyali veren hisseleri ekrana yazdryoruz
print("\nMACD AL sinyali veren hisseler:")
for sembol in al_sinyali_veren_hisseler:
    print(sembol)
```
