```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Coin'lerin fiyat verilerini ekme
def get_data(symbols, start_date, end_date):
    data = yf.download(symbols, start=start_date, end=end_date)['Adj Close']
    return data

# Coin'lerin BTC karsndaki performansn hesapla ve grafikle
def plot_btc_ratios(symbols, start_date, end_date):
    # Verileri ek
    data = get_data(symbols, start_date, end_date)
    data = data.dropna()

    # BTC fiyatn al
    btc_prices = data['BTC-USD']

    # Her bir coin iin BTC karsndaki oran hesapla
    plt.figure(figsize=(14, 10))
    for symbol in symbols:
        if symbol != 'BTC-USD':  # BTC'yi kendisiyle karlatrmayalm
            ratio = data[symbol] / btc_prices
            plt.plot(ratio, label=symbol)

    plt.title('Coinlerin BTC Karsndaki Performans')
    plt.xlabel('Tarih')
    plt.ylabel('BTC Karsndaki Oran')
    plt.legend()
    plt.grid()
    plt.show()

# rnek kullanm
symbols = ['BTC-USD', 'ETH-USD', 'BNB-USD', 'ADA-USD', 'DOGE-USD',
           'XRP-USD', 'DOT-USD', 'UNI-USD', 'LTC-USD', 'LINK-USD']
start_date = '2020-01-01'
end_date = '2023-01-01'

plot_btc_ratios(symbols, start_date, end_date)
```
