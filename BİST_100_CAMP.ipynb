```python
import yfinance as yf
import numpy as np
import pandas as pd

# Risksiz getiri oranı (örneğin, Türkiye Cumhuriyeti Devlet Tahvili getirisi)
risk_free_rate = 0.10  # %10 olarak varsayalım

# BIST 100 endeksi sembolü
market_symbol = "^XU100"  # BIST 100 endeksi (yfinance'de ^XU100 kullanılır)

# BIST 100 hisse senetleri sembolleri (örnek olarak birkaç hisse ekledim)
bist100_symbols = [
    "AEFES.IS", "AKBNK.IS", "AKSA.IS", "ALARK.IS", "ARCLK.IS", "ASELS.IS", "ASTOR.IS",
    "BAGFS.IS", "BERA.IS", "BIMAS.IS", "BIOEN.IS", "BRISA.IS", "BRYAT.IS", "BUCIM.IS",
    "CCOLA.IS", "CEMTS.IS", "CIMSA.IS", "DEVA.IS", "DOAS.IS", "DOHOL.IS", "ECILC.IS",
    "EGEEN.IS", "EKGYO.IS", "ENJSA.IS", "ENKAI.IS", "EREGL.IS", "EUPWR.IS", "FROTO.IS",
    "GARAN.IS", "GESAN.IS", "GUBRF.IS", "HALKB.IS", "HEKTS.IS", "ISCTR.IS", "ISFIN.IS",
    "ISMEN.IS", "IZMDC.IS", "KARSN.IS", "KARTN.IS", "KCAER.IS", "KCHOL.IS", "KLSYN.IS",
    "KONTR.IS", "KORDS.IS", "KOZAA.IS", "KOZAL.IS", "KRDMD.IS", "LOGO.IS", "MGROS.IS",
    "NETAS.IS", "ODAS.IS", "OTKAR.IS", "OYAKC.IS", "PETKM.IS", "PGSUS.IS", "QUAGR.IS",
    "SAHOL.IS", "SASA.IS", "SISE.IS", "SMRTG.IS", "SOKM.IS", "TAVHL.IS", "TCELL.IS",
    "THYAO.IS", "TKFEN.IS", "TOASO.IS", "TSKB.IS", "TTKOM.IS", "TUPRS.IS", "TURSG.IS",
    "ULKER.IS", "VAKBN.IS", "VESTL.IS", "YKBNK.IS", "YUNSA.IS", "ZOREN.IS"
]
# Veri aralığı
start_date = "2020-01-01"
end_date = "2023-01-01"

# Piyasa (BIST 100) verilerini çek
try:
    print(f"{market_symbol} endeksi verileri çekiliyor...")
    market = yf.Ticker(market_symbol)
    market_data = market.history(start=start_date, end=end_date)

    if market_data.empty:
        raise ValueError("Piyasa verileri çekilemedi. Lütfen sembolü ve tarih aralığını kontrol edin.")

    # Piyasa getirilerini hesapla
    market_returns = market_data['Close'].pct_change().dropna()

    # Sonuçları saklamak için bir liste oluştur
    results = []

    # Her bir hisse senedi için CAPM hesapla
    for symbol in bist100_symbols:
        try:
            print(f"{symbol} hissesi verileri çekiliyor...")
            stock = yf.Ticker(symbol)
            stock_data = stock.history(start=start_date, end=end_date)

            if stock_data.empty:
                print(f"{symbol} hissesi için veri bulunamadı.")
                continue

            # Hisse getirilerini hesapla
            stock_returns = stock_data['Close'].pct_change().dropna()

            # Hisse senedi ve piyasa getirilerini aynı uzunlukta hale getir
            min_length = min(len(stock_returns), len(market_returns))
            stock_returns = stock_returns[-min_length:]
            market_returns_aligned = market_returns[-min_length:]

            # Beta katsayısını hesapla
            covariance = np.cov(stock_returns, market_returns_aligned)[0, 1]
            market_variance = np.var(market_returns_aligned)
            beta = covariance / market_variance

            # Piyasa beklenen getirisi (örneğin, BIST 100'ün yıllık getirisi)
            market_expected_return = market_returns_aligned.mean() * 252  # Yıllıklaştırılmış getiri

            # CAPM formülü ile beklenen getiriyi hesapla
            expected_return = risk_free_rate + beta * (market_expected_return - risk_free_rate)

            # Sonuçları listeye ekle
            results.append({
                "Hisse": symbol,
                "Beta": beta,
                "Piyasa Beklenen Getirisi": market_expected_return,
                "Beklenen Getiri (CAPM)": expected_return
            })

        except Exception as e:
            print(f"{symbol} hissesi için hata oluştu: {e}")

    # Sonuçları DataFrame'e dönüştür
    results_df = pd.DataFrame(results)

    # Sonuçları Excel dosyasına kaydet
    output_file = "bist100_capm_results.xlsx"
    results_df.to_excel(output_file, index=False)
    print(f"Sonuçlar '{output_file}' dosyasına kaydedildi.")

except Exception as e:
    print(f"Hata oluştu: {e}")
```
