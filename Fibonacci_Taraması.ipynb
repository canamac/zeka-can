```python

!pip install git+https://github.com/rongardF/tvdatafeed
!pip install tradingview-screener


import pandas as pd
import numpy as np
from tvDatafeed import TvDatafeed, Interval
from tradingview_screener import get_all_symbols
import warnings
import time

warnings.simplefilter(action='ignore')

tv = TvDatafeed()

symbols = get_all_symbols(market='turkey')
symbols = [symbol.replace('BIST:', '') for symbol in symbols]
symbols = sorted(symbols)

print("Lütfen görmek istediğiniz zaman aralığını seçin (örneğin: 4 saatlik, günlük, haftalık, aylık):")
time_interval = input("Zaman aralığı: ").strip().lower()

interval_dict = {
    "4 saatlik": Interval.in_4_hour,
    "günlük": Interval.in_daily,
    "haftalık": Interval.in_weekly,
    "aylık": Interval.in_monthly
}
selected_interval = interval_dict.get(time_interval)

if selected_interval is None:
    print("Geçersiz zaman aralığı girdiniz. Varsayılan olarak günlük zaman aralığı kullanılacak.")
    selected_interval = Interval.in_daily

print("Lütfen görmek istediğiniz Fibonacci seviyelerini girin (örneğin: 0.0, 23.6, 38.2, 50.0, 61.8, 78.6, 100.0):")
fibonacci_levels_input = input("Seviyeler (virgülle ayrılmış): ")

try:
    fibonacci_levels = [float(level.strip()) for level in fibonacci_levels_input.split(",")]
except ValueError:
    print("Geçersiz Fibonacci seviyeleri girdiniz. Lütfen geçerli sayılar girin.")
    exit()

try:
    lookback = int(input("Kaç bar için tarama yapmak istiyorsunuz? (Varsayılan: 144): ").strip())
except ValueError:
    lookback = 144

def calculate_fibonacci_levels(df, levels, lookback=144):
    max_price = df['close'].tail(lookback).max()
    min_price = df['close'].tail(lookback).min()
    diff = max_price - min_price
    return {f"{level}%": max_price - (level / 100) * diff for level in levels}, max_price, min_price

print("Lütfen bekleyin, taramaya başlandı...")

for symbol in symbols:
    try:
        df = tv.get_hist(symbol=symbol, exchange='BIST', interval=selected_interval, n_bars=1000)

        if df.empty:
            print(f"{symbol} için veri çekilemedi.")
            continue

        fib_levels, max_price, min_price = calculate_fibonacci_levels(df, fibonacci_levels, lookback)

        latest_close = df['close'].iloc[-1]

        latest_low = df['low'].iloc[-1]
        latest_high = df['high'].iloc[-1]

        for label, level in fib_levels.items():
            if latest_low <= level <= latest_high:
                price_str = f"{level:.2f}".replace('.', ',')
                last_price = f"{latest_close:.2f}".replace('.', ',')

                if latest_close > level:
                    color_start = "\033[92m"
                    additional_info = " Fiyat, istenen Fibonacci seviyesinin üzerindedir."
                else:
                    color_start = "\033[91m"
                    additional_info = ""

                color_end = "\033[0m"

                print(f"{color_start}Hisse adı: {symbol}, Fiyat: {last_price}, "
                      f"İstenilen Fibonacci seviyesi {label}: {price_str}{additional_info}, "
                      f"Zaman aralığı: {time_interval}, Son {lookback} barda dip-tepe baz alınmıştır.{color_end}")

    except Exception as e:
        print(f"{symbol} için bir hata oluştu: {e}")
        continue

print("Tarama işlemi tamamlandı.")
print("Geri dönüşleriniz için X hesabım : https://x.com/yggdrasil_35")
```
