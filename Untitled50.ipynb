```python
function main() {
  /////////////// DEĞİŞTİRİLECEK KISIMLAR ///////////////
  // Sayfa Adı
  var sheetName = 'Bilanco';
  // Bilanco Donemleri
  var yil1= 2024
  var donem1 = 9
  var yil2= 2024
  var donem2 = 6
  var yil3= 2024
  var donem3 = 3
  var yil4= 2023
  var donem4 = 12
  var yil5= 2023
  var donem5 = 9
  var yil6= 2023
  var donem6= 6
  var yil7= 2023
  var donem7 = 3
  var yil8= 2022
  var donem8= 12
  // Hisse Secimi
  var hisse = "THYAO"
  /////////////// DEĞİŞTİRİLECEK KISIMLAR ///////////////


  // İlk 4 dönem çekme
  fetchData({
     hisse: hisse, yil1: yil1, donem1: donem1, yil2: yil2, donem2: donem2, yil3: yil3, donem3: donem3, yil4: yil4, donem4: donem4
  }, sheetName, 0);


  // Sonraki 4 dönem çekme
  fetchData({
    hisse: hisse, yil1: yil5, donem1: donem5, yil2: yil6, donem2: donem6, yil3: yil7, donem3: donem7, yil4: yil8, donem4: donem8},
    sheetName, 6); // Tabloyu doldurmaya 5. sütundan devam et
}


function fetchData(params, sheetName, columnOffset) {
  var {hisse, yil1, donem1, yil2, donem2, yil3, donem3, yil4, donem4} = params;


  var url = 'https://www.isyatirim.com.tr/_layouts/15/IsYatirim.Website/Common/Data.aspx/MaliTablo?companyCode=' + hisse + '&exchange=TRY&financialGroup=XI_29&year1=' + yil1 + '&period1=' + donem1 + '&year2=' + yil2 + '&period2=' + donem2 + '&year3=' + yil3 + '&period3=' + donem3 + '&year4=' + yil4 + '&period4=' + donem4;


  var response = UrlFetchApp.fetch(url);
  var json = JSON.parse(response.getContentText());


  if (json.ok) {
    var data = json.value;
    var sheet = SpreadsheetApp.getActiveSpreadsheet().getSheetByName(sheetName);


    // Sayfa yoksa ekleme
    if (!sheet) {
      sheet = SpreadsheetApp.getActiveSpreadsheet().insertSheet(sheetName);
    }


    // Eklenecek sütunlarin nereye eklenecegine karar verme
    // Sayfa bossa ilk sutuna ekle
    if (columnOffset === 0) {
      // Basliklari belirleme
      sheet.getRange(1, 1).setValue('Kod');
      sheet.getRange(1, 2).setValue('Tanım');
      columnOffset = 2;
    }


    // Basliklari belirleme
    sheet.getRange(1, columnOffset + 1).setValue(yil1 + '/' + donem1);
    sheet.getRange(1, columnOffset + 2).setValue(yil2 + '/' + donem2);
    sheet.getRange(1, columnOffset + 3).setValue(yil3 + '/' + donem3);
    sheet.getRange(1, columnOffset + 4).setValue(yil4 + '/' + donem4);


    // Verileri sayfaya ekleme
    for (var i = 0; i < data.length; i++) {
      var item = data[i];

      // Sadece ilk cagirmada Kod ve Tanım sutunlarini ekle
      if (columnOffset === 2) {
        sheet.getRange(i + 2, 1).setValue(item.itemCode);
        sheet.getRange(i + 2, 2).setValue(item.itemDescTr);
      }

      sheet.getRange(i + 2, columnOffset + 1).setValue(item.value1);
      sheet.getRange(i + 2, columnOffset + 2).setValue(item.value2);
      sheet.getRange(i + 2, columnOffset + 3).setValue(item.value3);
      sheet.getRange(i + 2, columnOffset + 4).setValue(item.value4);
    }
  } else {
    Logger.log("Error: " + json.errorDescription);
  }
}


// Ana fonksiyonu calistir
main();
```
