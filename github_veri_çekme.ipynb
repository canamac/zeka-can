```python
import requests

# GitHub'daki hedef dosyanın raw URL adresi
repo_url = "https://raw.githubusercontent.com/Onuragdaci/Trend_Channel/main/Trend_Channel.py"

# Raw içeriği requests ile çekme
response = requests.get(repo_url)

# Dosyayı Google Colab'a yazma ve çalıştırma
if response.status_code == 200:
    # İçeriği satırlara ayır
    lines = response.text.split('\n')

    # 'pip install' içeren tüm satırları kaldır
    code_lines = [line for line in lines if not line.strip().startswith('!pip install')]

    # Geri kalan içeriği birleştir
    code_to_run = '\n'.join(code_lines)

    # İçeriği exec ile çalıştır
    exec(code_to_run)
else:
    # Eğer hata oluştuysa, hatayı yazdır
    print(f"Error fetching the file: {response.status_code}")
```
