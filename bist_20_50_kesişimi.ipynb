```python
# Gerekli kütüphanelerin kurulumu
!pip install git+https://github.com/rongardF/tvdatafeed tradingview-screener
!pip install ta

import numpy as np
import pandas as pd
from tvDatafeed import TvDatafeed, Interval
from tradingview_screener import get_all_symbols
import ta
import warnings
import logging

warnings.simplefilter(action='ignore')

# Log ayarları
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()

# TvDatafeed başlatma
tv = TvDatafeed(username='your_username', password='your_password')

# EMA hesaplama fonksiyonu
def calculate_ema(data, period):
    return data.ewm(span=period, adjust=False).mean()

# EMA 20 ve EMA 50 kesişimi belirleme fonksiyonu
def check_ema_crossover(data):
    data['EMA20'] = calculate_ema(data['close'], 20)
    data['EMA50'] = calculate_ema(data['close'], 50)
    data['BullishCross'] = (data['EMA20'] > data['EMA50']) & (data['EMA20'].shift(1) <= data['EMA50'].shift(1))
    return data

# Türkiye piyasasından tüm sembolleri çekme
all_symbols = get_all_symbols(market='turkey')

# BIST sembollerini filtreleme
bist_symbols = [symbol for symbol in all_symbols if 'BIST:' in symbol]

# Taramayı gerçekleştirme
results = []

for symbol in bist_symbols:
    try:
        exchange, symbol_code = symbol.split(':')
        data = tv.get_hist(symbol=symbol_code, exchange=exchange, interval=Interval.in_daily, n_bars=100)
        data = data.reset_index()
        data = check_ema_crossover(data)

        last_row = data.iloc[-1]
        if last_row['BullishCross']:
            results.append({
                'Symbol': symbol,
                'Last Price': last_row['close'],
                'Date': last_row['datetime']
            })
        print(f"Processed {symbol}")
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")

# Sonuçları yazdırma
for result in results:
    print(f"Symbol: {result['Symbol']}, Last Price: {result['Last Price']}, Date: {result['Date']}")

```
